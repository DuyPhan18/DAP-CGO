```python
import pandas as pd
```

```python
data_path = 'C:\\Users\\duyph\\Documents\\DAP2308R1\\data\\'
path_name = 'OnlineRetail.csv'
df = pd.read_csv(data_path + path_name, encoding= 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
country = df['Country'].nunique()
print("Số lượng quốc gia: {}".format(country))
```

```python
df['total'] = df['UnitPrice'] * df['Quantity']
df.head()
```

```python
total_revenue = df['total'].sum()
print("Tổng doanh thu: {}".format(total_revenue))
```

```python
top_10 = df.groupby(['StockCode', 'Description'])['Quantity'].sum().sort_values(ascending= False).reset_index()
top_10.head(10)
```

```python
top_10_revenue = df.groupby(['StockCode', 'Description'])['total'].sum().sort_values(ascending= False).reset_index()
top_10_revenue.head(10)
```
